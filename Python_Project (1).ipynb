# 1. If-Else Basics

Example 1: Even or Odd

````python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
````

Lesson: if checks a condition, else runs when condition is false.

# Example 2: Age Check

````python
age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are NOT eligible to vote")
````

Lesson: multiple outcomes depending on condition. Students can play with different ages.

# Example 3: Using elif

````python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
````

Lesson: multiple decisions with elif.

# 2. While Loops

## Example 1: Print Numbers 1â€“5

````python
i = 1
while i <= 5:
    print("Number:", i)
    i += 1
````

# Example 2: Password Check

````python
password = "lpu123"
attempt = ""

while attempt != password:
    attempt = input("Enter password: ")

print("Access Granted")
````

Lesson: loop continues until correct password is entered.

# Example 3: Countdown

````python
n = 5
while n > 0:
    print(n)
    n -= 1
print("Blast Off")
````

Lesson: decrementing loop with stop condition.

## Mini Demo: Menu Program with If-Else, While, and Break

````python
# Simple Menu Program

while True:   # loop will keep running until we "break"
    print("\n===== Simple Menu =====")
    print("1. Greet")
    print("2. Add Two Numbers")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        print("Hello, Welcome to Python Learning! ðŸ‘‹")

    elif choice == "2":
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))
        print("Sum is:", a + b)

    elif choice == "3":
        print("Exiting program... Goodbye! ðŸ‘‹")
        break   # stops the loop immediately

    else:
        print("Invalid choice. Please select 1, 2, or 3.")
````
